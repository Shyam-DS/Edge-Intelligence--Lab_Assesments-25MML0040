```python
import nltk
```

```python
def gender_features(words):
    return{'last letter': words[-1]}
```

```python
nltk.download('names')
```

```python
from nltk.corpus import names
names.words()
print(len(names.words()))
```

```python
labeled = ([(name,'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled)
```

```python
feature_sets = [(gender_features(n),gender) for (n,gender) in labeled]
```

```python
train_set , test_set = feature_sets[:5000], feature_sets[:2000]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('David'))
classifier.classify(gender_features('Yann Lecunn'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```
